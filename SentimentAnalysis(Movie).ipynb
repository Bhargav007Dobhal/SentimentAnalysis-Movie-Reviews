```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('/content/IMDB_Dataset.csv')
```

```python
df['sentiment'].replace({'positive':1,'negative':0},inplace=True)
```

```python
df.head()
```

```python
#Removing html tags
#Importing the regex or you can say Regular Expression library to do this task
import re
#clean=re.compile('<.*?>') Creates a stored pattern that identifies HTML tags. (Specifically: finds anything starting with < and ending with the first > it sees).
clean=re.compile('<.*?>')
# re.sub(clean,'',df.iloc[2].review) Scans the text in the 3rd row of the review column, deletes any HTML tags found (replaces them with nothing), and returns the clean text.
re.sub(clean,'',df.iloc[2].review)
```

```python
def clean_html(text):
  clean=re.compile('<.*?>')
  return re.sub(clean,' ',text)
```

```python
df['review']=df['review'].apply(clean_html)
df.head()
```

```python
def convert_lower(text):
  return text.lower()
```

```python
#Converting every letter of the review to lower case
df['review']=df['review'].apply(convert_lower)
df.head()
```

```python
#Function to remove special character
def remove_special(text):
    X=" "
    for i in text:
        if i.isalnum():
            X=X+i
        else:
            X=X+' '
    return X
```

```python
#Removing special characters from the reviews
df['review']=df['review'].apply(remove_special)
df.head()
```

```python
#Removing stop words
import nltk
from nltk.corpus import stopwords
```

```python
df.head()
```

```python
#To check all the stop words in english language you can write this code
#By using len function we can say that there are a total of 198 stop words in english
import nltk
from nltk.corpus import stopwords

# 1. Download the resource (only needed once, but safe to leave in)
nltk.download('stopwords')

# 2. Your original code
# To check all the stop words in english language you can write this code
print(stopwords.words('english'))

# By using len function we can say that there are a total of 179 (varies by version) stop words
print(len(stopwords.words('english')))
```

```python
#Function to remove stop words
def remove_stopwords(text):
    X=[]
    # splitting each review into a list of words and then checking if that word is present in the stopwords for english if not we are appending them to an empty list.
    for i in text.split():
        if i not in stopwords.words('english'):
            X.append(i)
    Y=X[:]
    X.clear()
    return Y
```

```python
df['review']=df['review'].apply(remove_stopwords)
```

```python
#last step of text preprocessing stemming
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
#This function does stemming
output=[]
z=[]
def stem_words(text):
    for i in text:
        output.append(ps.stem(i))
    z=output[:]
    output.clear()
    return z
```

```python
#Example of how stemming works
stem_words(['I','Loved','Loving','it'])
```

```python
df['review']=df['review'].apply(stem_words)
```

```python
#converting list back to a string
def join_back(list_input):
    return " ".join(list_input)
```

```python
#Converting the list back to a string
df['review']=df['review'].apply(join_back)
```

```python
#Using count vectorizer to convert each unique words including all the reviews into columns
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000)
```

```python
X=cv.fit_transform(df['review']).toarray()
```

```python
import os
import threading
import time
import subprocess
import requests
from flask import Flask, request, jsonify
from flask_cors import CORS

# --- STEP 1: FORCE STOP OLD SERVERS ---
print("Stopping old servers...")
os.system("fuser -k 5000/tcp")

# --- STEP 2: DOWNLOAD CLOUDFLARE TUNNEL ---
if not os.path.exists("cloudflared-linux-amd64"):
    print("Downloading Cloudflare Tunnel...")
    os.system("wget -q https://github.com/cloudflare/cloudflared/releases/latest/download/cloudflared-linux-amd64")
    os.system("chmod +x cloudflared-linux-amd64")

# --- STEP 3: START FLASK ---
app = Flask(__name__)
CORS(app)

@app.route('/predict', methods=['POST'])
def predict():
    data = request.json
    review_text = data.get('text', '')

    # Helper function (needed because 'ps' is global)
    def stem_words_safe(text):
        output = []
        for i in text:
            output.append(ps.stem(i))
        return output

    try:
        # Preprocessing (Uses functions defined in Cells 1-49)
        cleaned = clean_html(review_text)
        lowered = convert_lower(cleaned)
        no_special = remove_special(lowered)
        no_stopwords = remove_stopwords(no_special)
        stemmed = stem_words_safe(no_stopwords)
        final_text = join_back(stemmed)

        # Predict (Uses 'cv' and 'clf3' trained in Cells 1-49)
        vectorized_text = cv.transform([final_text]).toarray()
        prediction = clf3.predict(vectorized_text)

        result = "Positive" if prediction[0] == 1 else "Negative"
        return jsonify({'sentiment': result})

    except Exception as e:
        return jsonify({'error': str(e)})

def run_app():
    app.run(port=5000)

threading.Thread(target=run_app).start()

# --- STEP 4: START CLOUDFLARE TUNNEL ---
print("\n Starting Cloudflare Tunnel...")
time.sleep(2)

logfile = open('cloudflared.log', 'w')
process = subprocess.Popen(['./cloudflared-linux-amd64', 'tunnel', '--url', 'http://127.0.0.1:5000'], stdout=logfile, stderr=logfile)

print("Waiting for URL to generate...")
time.sleep(5)

found_url = False
with open('cloudflared.log', 'r') as f:
    for line in f:
        if '.trycloudflare.com' in line:
            url = "https://" + line.split("https://")[1].split()[0]
            print("\n SUCCESS! YOUR BACKEND URL IS:")
            print(f" {url}")
            found_url = True
            break

if not found_url:
    print("\n URL not found yet. Run '!cat cloudflared.log' in a new cell to find it.")
```

```python
# @title  Movie Review Sentiment Analyzer
import ipywidgets as widgets
from IPython.display import display, clear_output
import requests
import json

# ------------------------------------------------------------------
# PASTE YOUR CLOUDFLARE URL BELOW (Keep the quotes!)
# Example: url = "https://boats-brooks-gardens-toys.trycloudflare.com"
url = "https://testimonials-meyer-holidays-assumed.trycloudflare.com"
# ------------------------------------------------------------------

# Ensure the URL ends with /predict
if not url.endswith('/predict'):
    url = url.rstrip('/') + "/predict"

# Create the text input box
text_input = widgets.Textarea(
    value='',
    placeholder='Type your movie review here... (e.g., "I loved this movie, it was fantastic!")',
    description='Review:',
    disabled=False,
    layout=widgets.Layout(width='600px', height='100px')
)

# Create the button
button = widgets.Button(
    description='Analyze Sentiment',
    disabled=False,
    button_style='success', # 'success', 'info', 'warning', 'danger' or ''
    tooltip='Click to predict',
    icon='check'
)

# Output area for the result
output = widgets.Output()

def on_button_clicked(b):
    with output:
        clear_output()
        review = text_input.value
        if not review:
            print(" Please enter some text first.")
            return

        print(" Analyzing...")

        try:
            # Send the text to your running Flask server
            response = requests.post(url, json={'text': review})

            if response.status_code == 200:
                result = response.json()
                sentiment = result.get('sentiment', 'Unknown')

                # Display result with formatting
                if sentiment == "Positive":
                    print(f"\n RESULT: {sentiment} ")
                else:
                    print(f"\n RESULT: {sentiment} ")
            else:
                print(f"Error: Server returned status code {response.status_code}")

        except Exception as e:
            print(f"\n Connection Error: Could not connect to your server.")
            print(f"Make sure the cell with 'Cloudflare Tunnel' is still running!")
            print(f"Error details: {e}")

button.on_click(on_button_clicked)

# Display everything
display(text_input, button, output)
```
